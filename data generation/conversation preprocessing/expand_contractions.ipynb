```python
# Import necessary libraries
import contractions
```

```python
def read_conversations(file_path):
    """
    Read conversations from a file.

    """
    with open(file_path, 'r', encoding='utf-8') as file:
        data = file.read()
    return data


def expand_contractions(data):
    """
    Expand contractions in each conversation.
    
    """
    # Split the input data into separate conversations based on two newlines
    conversations = data.strip().split('\n\n')
    
    # Expand contractions in each conversation
    expanded_conversations = [contractions.fix(conversation) for conversation in conversations]
    
    # Join expanded conversations back with two newlines
    return '\n\n'.join(expanded_conversations)


def save_conversations(expanded_conversations, output_file_path):
    """
    Save expanded conversations to a file.
    
    """
    with open(output_file_path, 'w', encoding='utf-8') as file:
        file.write(expanded_conversations)


def main(input_file_path, output_file_path):
    """
    Main function to read, expand, and save conversations.
    
    """
    # Read data from the file
    data = read_conversations(input_file_path)
    
    # Expand conversations
    expanded_conversations = expand_contractions(data)
    
    # Save the expanded conversations to the output file
    save_conversations(expanded_conversations, output_file_path)


# Paths to the old and new files
files_paths = {
    "train": ("data preprocessing/data/initial_train_data.txt",
              "data preprocessing/data/preprocessed_train_data.txt"),
    "validation": ("data preprocessing/data/initial_validation_data.txt",
                   "data preprocessing/data/preprocessed_validation_data.txt"),
    "test": ("data preprocessing/data/initial_test_data.txt",
             "data preprocessing/data/preprocessed_test_data.txt")
}

# Process and save all data
for key, paths in files_paths.items():
    main(*paths)
```
