```python
import pandas as pd
import re
```

```python
# Read data
def read_conversations(file_path):

    names = ['Mohammed', 'Abdullah', 'Rudolf', 'Johan', 'Ali', 'Mehmet', 'Pieter', 'Jan', 'Aicha', 'Fatima', 'Julia', 'Ingrid', 'Ayşe', 'Fatma', 'Johanna', 'Maria']

    with open(file_path, 'r', encoding='utf-8') as file:
        content = file.read()
        conversations = content.strip().split('\n\n')  

    data = []
    for convo in conversations:
        convo_name = None
        for name in names:
            # Check if the name appears in the conversation
            if re.search(r'\b' + re.escape(name) + r'\b', convo):
                convo_name = name
                break  # Stop after the first match

        initiator = 'Patient' if convo.strip().startswith('P:') else 'Agent'
        # Append data only if a name is found
        if convo_name:
            data.append({'conversation': convo, 'name': convo_name, 'initiator': initiator})

    return pd.DataFrame(data)

# Load data
file_path = 'C:/Users/ntanavarass/Desktop/Conversational-Triple-Extraction/data generation/conversations generation/raw conversations/few_shot_raw_conversations_gpt4.txt'
df = read_conversations(file_path)
print(df.head(100))  
```

```python
train_set = []
validation_set = []
test_set = []

# Group data by persona name and distribute conversations
grouped = df.groupby('name')
for name, group in grouped:
    
    print(f"Processing {name} with {len(group)} conversations.")  # Debugging line
   
    train_conversations = group.iloc[:10]
    validation_conversations = group.iloc[10:13]
    test_conversations = group.iloc[13:]

    # Append to the respective set
    train_set.append(train_conversations)
    validation_set.append(validation_conversations)
    test_set.append(test_conversations)

# Function to format data as plain text
def format_as_text(df):
    return '\n\n'.join(df['conversation'])

# Concatenate all persona-specific frames into single frames for each set
train_df = pd.concat(train_set).reset_index(drop=True)
validation_df = pd.concat(validation_set).reset_index(drop=True)
test_df = pd.concat(test_set).reset_index(drop=True)

# Format data
train_text = format_as_text(train_df)
validation_text = format_as_text(validation_df)
test_text = format_as_text(test_df)


print('Train set size:', len(train_df))
print('Validation set size:', len(validation_df))
print('Test set size:', len(test_df))


# Save the datasets
with open('C:/Users/ntanavarass/Desktop/Conversational-Triple-Extraction/data preprocessing/data/initial_train_data.txt', 'w', encoding='utf-8') as file:
    file.write(train_text)
with open('C:/Users/ntanavarass/Desktop/Conversational-Triple-Extraction/data preprocessing/data/initial_validation_data.txt', 'w', encoding='utf-8') as file:
    file.write(validation_text)
with open('C:/Users/ntanavarass/Desktop/Conversational-Triple-Extraction/data preprocessing/data/initial_test_data.txt', 'w', encoding='utf-8') as file:
    file.write(test_text)
```
